```python
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense

# Generate training data
X = np.linspace(-np.pi, np.pi, 200)
y = np.sin(X)

# Build the neural network
model = Sequential()
model.add(Dense(10, input_dim=1, activation='sigmoid'))
model.add(Dense(1, activation='linear'))
model.compile(loss='mse', optimizer='adam')

# Train the neural network
model.fit(X, y, epochs=1000, verbose=0)

# Evaluate the neural network
test_X = np.linspace(-np.pi, np.pi, 200)
test_y = np.sin(test_X)
predictions = model.predict(test_X)
mse = np.mean((test_y - predictions)**2)

# Plot the results
plt.plot(test_X, test_y, label='Ground truth')
plt.plot(test_X, predictions, label='Predictions')
plt.title(f'Sine Function Approximation (MSE={mse:.4f})')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense

# Get the function from the user
func_str = input('Enter a mathematical function of one variable: ')
x_str = input('Enter the variable name: ')
x = np.linspace(-10, 10, 200)
y = eval(func_str)

# Build the neural network
model = Sequential()
model.add(Dense(10, input_dim=1, activation='sigmoid'))
model.add(Dense(1, activation='linear'))
model.compile(loss='mse', optimizer='adam')

# Train the neural network
model.fit(x, y, epochs=10000, verbose=0)

# Evaluate the neural network
test_x = np.linspace(-10, 10, 200)
test_y = eval(func_str.replace(x_str, 'test_x'))
predictions = model.predict(test_x)
mse = np.mean((test_y - predictions)**2)

# Plot the results
plt.plot(test_x, test_y, label='Ground truth')
plt.plot(test_x, predictions, label='Predictions')
plt.title(f'{func_str} Approximation (MSE={mse:.4f})')
plt.legend()
plt.show()
```
